### select period to fit factor model

Notice: different window_size or start_date for same security gets same result, see detailed example below. Some sec doesn't fit well

```python
start_date = "2009-01-01"
window_size = 250*3
mdl = {}
summary = {}
for sec in ["EMB","EDV","TFI","JJM","DBB","PCY"]:
    mdl[sec] = universe.fitFactorModel(sec, start_date, window_size)
    summary[sec] = mdl[sec].summary()

# a list of low R-squared
# "EMB" 0.388 "EDV" 0.543 TFI 0.359 JJM 0.5 DBB 0.507 PCY 0.427
```

```python
start_date = "2010-01-01"
window_size = 250*3
mdl2 = {}
summary2 = {}
for sec in ["EMB","EDV","TFI","JJM","DBB","PCY"]:
    mdl2[sec] = universe.fitFactorModel(sec, start_date, window_size)
    summary2[sec] = mdl[sec].summary()
```
